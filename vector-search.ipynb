```python
import os
import requests
from dotenv import load_dotenv
from typing import List
from langchain_core.embeddings import Embeddings
from langchain_core.documents import Document
from langchain.vectorstores import Chroma

# Load Gemini API key from .env
load_dotenv()
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

# Gemini embedding endpoint
EMBEDDING_ENDPOINT = "https://generativelanguage.googleapis.com/v1beta/models/embedding-001:embedContent"

# ðŸ”§ Custom embedding class
class GeminiEmbeddings(Embeddings):
    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        return [self._embed(text) for text in texts]

    def embed_query(self, text: str) -> List[float]:
        return self._embed(text)

    def _embed(self, text: str) -> List[float]:
        headers = {"Content-Type": "application/json"}
        params = {"key": GEMINI_API_KEY}
        data = {
            "content": {
                "parts": [{"text": text}]
            }
        }
        response = requests.post(EMBEDDING_ENDPOINT, headers=headers, params=params, json=data)
        response.raise_for_status()
        return response.json()["embedding"]["values"]

# âœ… Sample documents
texts = [
    "A delightful story about birds, trees, and rivers for kids.",
    "Learn nature with interactive illustrations and poems.",
    "A children's guide to ecosystems and wildlife."
]
documents = [Document(page_content=text) for text in texts]

# âœ… Initialize Chroma with custom Gemini embeddings
embedding = GeminiEmbeddings()
db_books = Chroma.from_documents(documents, embedding=embedding, collection_name="books")

# âœ… Query
query = "A book to teach children about nature"
results = db_books.similarity_search(query, k=3)

# âœ… Print results
for i, doc in enumerate(results, 1):
    print(f"\nðŸ“š Result {i}:\n{doc.page_content}")
```
